```
import os
import numpy as np

#Make a numpy array where each row contains image_path and details of each face in the image
def read_images(base_dir,image_list):
  all = []
  count = 0
  y = []
  for txt in image_list:
    path = base_dir +'/'+ txt.split('P')[0]
    with open(os.path.join(base_dir,txt),'r') as f:
        lines = f.readlines()[0:]
        for line in lines:
            data = line.strip().split('\n')
            if '.jpg' in data[0]:
              pz = y.copy()
              all.append(pz)
              y.clear()
              x = path + data[0]
              y.append(x)
              count= count+1  
            else:
              z = data[0].strip().split('\t')
              z.append(False)
              y.append(z)
  print(y)
  all.append(y)
  all = np.array(all)
  all = all[1:]
  print("Count:",count)
  return all

def get_images():
  base_dir = '/content/drive/My Drive/Gallagher'
  image_list = ['Fam2a/PersonData.txt'
             ,'Fam4a/PersonData.txt','Fam5a/PersonData.txt','Fam8a/PersonData.txt'
              ,'Group2a/PersonData.txt','Group4a/PersonData.txt','Group5a/PersonData.txt',
              'Group8a/PersonData.txt','Wed2a/PersonData.txt','Wed3a/PersonData.txt','Wed5a/PersonData.txt']

  all_images_array = read_images(base_dir,image_list)
  return all_images_array
 
all_images = get_images()
```

```
#Get single faces from the group images
import cv2
import time
from google.colab.patches import cv2_imshow
os.chdir("/content/drive/My Drive/")
data = []
casc = '/content/drive/My Drive/haarcascade_frontalface_default.xml'
faceCascade = cv2.CascadeClassifier(casc)
count = 0
start_time = time.time()
#path_val = "image_data/val/"
#path_train = "image_data/train/"
#val_count = 5237
for image in all_images:
  image_path = image[0]
  total_faces = image[1:]
  img = cv2.imread(image_path)
  gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) 
  faces = faceCascade.detectMultiScale(
            img,
            scaleFactor=1.045,
            minNeighbors=3
  )
  rect_pts = []
  for (x,y,w,h) in faces:
    rect_pts.append([x,y,w,h])
  rect = len(rect_pts)
  for face in total_faces:
    left_x = int(face[0])
    left_y = int(face[1])
    right_x = int(face[2])
    right_y = int(face[3])
    age = int(face[4])
    gender = int(face[5])
    if gender == 2:
      gender = 0
    for i in range(rect):
      max_x = rect_pts[i][0] + rect_pts[i][2]
      max_y = rect_pts[i][1] + rect_pts[i][3]
      min_x = rect_pts[i][0]
      min_y = rect_pts[i][1]
      if left_x >= min_x and left_x < max_x:
         if right_y >= min_y and left_y < max_y:
            if face[6] == True:
              continue        
            else:
              face[6] = True
              h = max_y - min_y
              w = max_x - min_x
              face_img = img[min_y:min_y+h,min_x:min_x+w]
              face_img  = cv2.resize(face_img,(224,224))
              face_img = face_img.reshape(1, -1)[0]
              #print(type(face_img))
              #print(face_img.shape)
              z = []
              z.append(face_img)
              #print(face_img)
              print("Face:",face_img[0])
              z.append(age)
              z.append(gender)
              data.append(z)
              #print(z)
              count += 1
              print(count)
              #if count < val_count:
              #  path = path_val + str(gender)
              #  path = path +'/'
              #  path = path + str(age)
              #  image_name = str(count)
              #  image_name = image_name + '.jpg'
                #cv2.imwrite(os.path.join(path ,image_name), face_img)    
              #else:
              #  path = path_train + str(gender)
              #  path = path +'/'
              #  path = path + str(age)
              #  image_name = str(count)
              #  image_name = image_name + '.jpg'
                #cv2.imwrite(os.path.join(path ,image_name), face_img)

                
end_time = time.time()
print(count)
print("Time taken:",end_time - start_time)
print(data)
```
